```python
import pymongo
from pymongo import MongoClient
```

```python
client = MongoClient('localhost', 27017)
db = client["softwaredb"]
db
```

```python
# Perform a join using $lookup
results = db.numeric.aggregate( [
   {
     '$lookup':
       {
         'from': 'text',
         'localField': 'reviewNumber',
         'foreignField': 'reviewNumber',
         'as': 'combined'
       }
  }
] )
```

```python
# print lookup results
for item in results:
    print(item)
```
